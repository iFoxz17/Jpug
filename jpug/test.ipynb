## Importazione delle librerie

```python
from encoder.RGB_Encoder import RGB_Encoder
from encoder.L_Encoder import L_Encoder

from Parser import Parser

from PIL import Image
```

## Caricamento dell'immagine

```python
bmp_image_path = "..\\examples\\silicon_valley.bmp"

image = Image.open(bmp_image_path)

image
```

## Gray-scaled Encoder

### Creazione encoder

```python
F = 8
d = 9

l_encoder = L_Encoder(F, d)
l_encoder
```

### Codifica

```python
jpug = l_encoder.encode(image)
jpug
```

### Salvataggio

```python
file_path = '..\examples\silicon_valley.jpug'
Parser.save(jpug, file_path)
```

### Caricamento

```python
jpug = Parser.load(file_path)
jpug
```

### Decodifica

```python
compressed_image = l_encoder.decode(jpug)
compressed_image
```
